```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dados_vendas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dados_vendas')
dados_vendas.head()
```

```python
ibge_sp = pd.read_csv(
    'https://gist.githubusercontent.com/esaheki/dedc519545db1ce5911174e8fcec8515/raw/acaafd5ecc45bbf70617ba36893a031ce9aa1d28/Basico_SP1.csv',
    encoding='utf8',
    sep=';', thousands='.', decimal=','
)
ibge_sp.dropna(how='all', axis=1, inplace=True)
ibge_sp.head()
```

```python
ibge_sp.info()
```

```python
enderecos = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/enderecos.csv')
enderecos.head()
```

```python
#endereçando somente os dados do estado de sp
enderecos_sp = enderecos.query("sigla_uf == 'SP'")
enderecos_sp.head()
```

```python
#ajustando os dados 
enderecos_sp['rua'] = enderecos_sp['tipo_logr']+ " " + enderecos_sp['logr_nome']
enderecos_sp['rua'] = enderecos_sp['rua'].str.lower().str.strip()
enderecos.head()
```

```python
dados_vendas["Rua"].str.extract(r'(^[\w ]+)')[:10]
```

```python
dados_vendas['apenas_rua'] = dados_vendas['Rua'].str.extract(r'(^[\w ]+)')
dados_vendas['apenas_rua'] = dados_vendas['apenas_rua'].str.lower().str.strip()
dados_vendas.head()
```

Utilizando metodo merge no padas

```python
dados_geo = pd.merge(left = dados_vendas, right = enderecos_sp[['rua', 'cep', 'latitude', 'longitude']], how = 'left', left_on = 'apenas_rua', right_on = 'rua').drop_duplicates(subset=dados_vendas.columns).query('cep > 0')
dados_geo
```
